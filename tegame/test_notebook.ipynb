```python
from copy import deepcopy
hand=[40,41,50,51,70,1,2,3,20,30,31]
hand_ = deepcopy(hand)
hand_.sort()

pairs=[]
for i, card_i in enumerate(hand_):
    combo = []
    combo.append(i)
    old = card_i
    if card_i in pairs: continue
    for j, card_j in enumerate(hand_):
        if card_j - old == 10:
            combo.append(j)
            old = card_j
    if is_empty_lists(pairs):
        if (len(combo) > 1) : pairs.append(combo)
    else:
        if (len(combo) > 1) and not is_list_of(pairs,combo) : pairs.append(combo)
pairs
```
