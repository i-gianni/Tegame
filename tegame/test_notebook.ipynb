```python
from copy import deepcopy
hand=[40,41,50,51,70,1,2,3,20,30,31]
hand_ = deepcopy(hand)
hand_.sort()

pairs=[]
for i, card_i in enumerate(hand_):
    combo = []
    combo.append(i)
    old = card_i
    if card_i in pairs: continue
    for j, card_j in enumerate(hand_):
        if card_j - old == 10:
            combo.append(j)
            old = card_j
    if is_empty_lists(pairs):
        if (len(combo) > 1) : pairs.append(combo)
    else:
        if (len(combo) > 1) and not is_list_of(pairs,combo) : pairs.append(combo)
pairs
```

```python
def argsort(seq):
    """Return the indices that would sort the sequence."""
    return [i for i, _ in sorted(enumerate(seq), key=lambda x: x[1])]

# Example usage
list_a = [3, 1, 4, 2]
list_b = ['c', 'a', 'd', 'b']

sorted_indices = argsort(list_a)
#print(sorted_indices)  # [1, 3, 0, 2]

# Apply the indices to rearrange both lists
list_a_sorted = [list_a[i] for i in sorted_indices]
list_b_rearranged = [list_b[i] for i in sorted_indices]

list_a_orig = [list_a_sorted[i-1] for i in list_a]
print(list_a_orig)

#print(list_a_sorted)    # [1, 2, 3, 4]
#print(list_b_rearranged)  # ['a', 'b', 'c', 'd']
```
