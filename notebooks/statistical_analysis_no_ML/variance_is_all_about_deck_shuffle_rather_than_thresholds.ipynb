```python
import pandas as pd
from tqdm import tqdm

import sys
sys.path.append('../')
from tegame import *

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Parametri
threshold1_range = range(1, 10)
threshold2_range = range(1, 20)
N_states = 300

# 1. Genera molti stati iniziali
states = []
for _ in range(N_states):
    g = Tegame(players=2)
    g.restart()
    states.append(g.get_state().copy())

# 2. Varianza tra stati (threshold fisso)
fixed_thresh = (10, 5)
state_results = []

for idx in range(N_states):
    game = Tegame(
        players=2,
        verb_lvl=0,
        thresh_secondchoice=fixed_thresh[0],
        thresh_nonmandatory=fixed_thresh[1]
    )
    game.set_state(states[idx].copy())
    state_results.append(int(game.run_game()))

state_variance = np.var(state_results)

# 3. Varianza tra threshold (stato fisso)
fixed_state = states[0].copy()
threshold_results = []

for thresh_sec in threshold2_range:
    for thresh_non in threshold1_range:
        game = Tegame(
            players=2,
            verb_lvl=0,
            thresh_secondchoice=thresh_sec,
            thresh_nonmandatory=thresh_non
        )
        game.set_state(fixed_state.copy())
        threshold_results.append(int(game.run_game()))

threshold_variance = np.var(threshold_results)

# 4. Rapporto tra varianze
variance_ratio = state_variance / (threshold_variance + 1e-12)  # evita divisione per zero

# 5. Stampa con label chiare
print("=== Variance Analysis ===")
print(f"1. Variance across states (state variance):       {state_variance:.6f}")
print(f"2. Variance across thresholds (threshold variance): {threshold_variance:.6f}")
```
